```python
from bokeh.io import output_notebook, show
from bokeh.layouts import column, row
from bokeh.models import Select, Slider, FactorRange
from bokeh.plotting import figure
import numpy as np
import random

# Enable Jupyter Notebook integration
output_notebook()

# Create the first plot with modified colors and reduced size
fig1 = figure(title="Polygon Plot", width=300, height=200)
x = [[[[0, 0, 1, 1]]], [[[2, 2, 4, 4], [2.5, 2.5, 3.5, 3.5]]], [[[2, 0, 4]]]]
y = [[[[2.5, 0.5, 0.5, 2.5]]], [[[1, 0, 0, 1], [0.75, 0.25, 0.25, 0.75]]], [[[2, 0, 0]]]]
color = ["red", "blue", "green"]
fill_alpha = 0.5
fig1.multi_polygons(x, y, color=color, fill_alpha=fill_alpha)

# Create the second plot with modified colors and reduced size
fig2 = figure(title="Scatter Plot", width=300, height=200)
x = np.arange(5)
y = x ** 2
z = x * 3
p = np.linspace(1, 20, 7)
q = np.linspace(1, 10, 7)
r = np.linspace(1, 30, 5)
a = np.arange(31)
fig2.circle(x, y, color="orange", size=10)
fig2.square(x, z, color="purple", size=7, alpha=0.4)
fig2.hex(y, z, color="green", size=5, alpha=0.6)
fig2.line(x, y, color="gray", line_width=1, line_dash="dotted")
fig2.inverted_triangle(p, q, color="brown", size=10, alpha=0.3)
fig2.diamond(x, r, color="teal", size=8, alpha=0.5)
fig2.cross(a, a, size=7, color="black")

# Create the third plot with modified colors and reduced size
fig3 = figure(title="Line Plot", width=300, height=200)
x = []
y = []
for i in range(100):
    x.append(i)
for i in range(100):
    y.append(1 + random.random())
line_color = "blue"
line_dash = "dashed"
legend_label = "Line 1"
fig3.line(x, y, line_color=line_color, line_dash=line_dash, legend_label=legend_label)

# Add a second line to the third plot
x = []
y = []
for i in range(100):
    x.append(i)
for i in range(100):
    y.append(random.random())
line_color = "green"
line_dash = "dotdash"
line_dash_offset = 1
legend_label = "Line 2"
fig3.line(x, y, line_color=line_color, line_dash=line_dash, line_dash_offset=line_dash_offset, legend_label=legend_label)

# Create the fourth plot with reduced size
x = np.linspace(0, 10, 100)
y = np.sin(x)
fig4 = figure(title="Sine Curve Plot", width=300, height=200)
fig4.line(x, y, color="purple")

# Create the layout for the four plots
layout = column(row(fig1, fig2), row(fig3, fig4))

# Display the layout in the Jupyter Notebook
show(layout)
```
